```python
def euc_dis(data1,data2):
#     print("hello")
    dist=0
    dist=(np.square(data1['sepal_length']- data2['sepal_length']) + np.square(data1['sepal_width']- data2['sepal_width']) + np.square(data1['petal_length']- data2['petal_length'])+ np.square(data1['petal_width']- data2['petal_width']))
    dist=np.sqrt(dist)
    return dist
    
    
```

```python
test_ans=[]
k=int(input("enter the value of k"))
for j in range(0,len(test)):
    diste=[]
    for i in range(0,len(train_x)):
        x=euc_dis(train_x.loc[i],test.iloc[j])
        diste.append(x)
    diste.sort()
    mean=0
    for ks in range(0,k):
        mean=mean+diste[ks]
    mean=mean/k
    test_ans.append(mean)
test_ans
```
